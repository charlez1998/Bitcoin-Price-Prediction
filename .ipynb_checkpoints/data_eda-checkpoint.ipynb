## Testing Stationarity

Before applying any model to our time series, we have to understand how it behaves and determine whether there is a trend or not. 

In this case, we will apply an Augmented Dicky Fuller test. It determines how strongly a time series is defined by a trend.

Hence our hypothesis will be conducted in the following way:

Null Hypothesis (H0): Null hypothesis of the test is that the time series can be represented by a unit root that is not stationary.
Alternative Hypothesis (H1): Alternative Hypothesis of the test is that the time series is stationary.

```python
from statsmodels.tsa.stattools import adfuller
```
